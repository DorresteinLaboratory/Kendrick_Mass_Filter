## Load Data Matrix

```r
# PERFORM: change path to .csv
#    Load in data matrix created in MZmine
#    Transpose matrix to get into the correct format for subsequent processes
#    Only keeps the (mass-to-charge;retention time) information
    data_matrix <- read.csv("all17K/tab17swabextract.csv", header=FALSE)
    #data_matrix <- read.table("20170731_kendrickmass_orbi_transposed_filterthis.csv", header=FALSE, sep="\t")
    mat1data <- data_matrix
    mat1data <- mat1data[1,]
#    Turn string vector into matrix
    mat1data <- as.matrix(mat1data)
    mat1data <- t(mat1data)
#    Transpose matrix to get into the correct format for subsequent processes
#    Retain the (mass-to-charge;retention time) and peak area information
    mat2data <- data_matrix
#    double-check values
#        check mat1data contains the (mass-to-charge;retention time) variables
#        check mat2data contains the (mass-to-charge;retention time) variables and associated peak areas
    head(mat1data)
    head(mat2data)
    dim(mat1data)
    dim(mat2data)
```

```r
# PERFORM: run cell
#    split string which contains concat m/z and retention time information
#    check size of the split string
    vec <- t(sapply(mat1data[-1,1], function(x) strsplit(as.character(x), ";")[[1]]))
#    make values in vector numeric 
    vec <- apply(vec, 2, as.numeric)
#    check first few that split was performed correctly    
    dim(vec)
    head(vec)
```

## Calculate All Possible Feature Combinations

```r
# PERFORM: run cell
#    recal the previously computed size of vector of m/z and RT - check for the # of variables
    dim(vec)
#    check dimensions of new matrix "cb" - this value should be large
    cb <- combn(1:(length(vec[,1])), 2)
    dim(cb)
```

## Select Desired Filter

```r
# PERFORM: choose polymer (or other) by uncommenting line and run cell
#    calculate the value to scale the m/z from IUPAC scale to kendrick mass scale (scale values by unit repeat: integer mass/monoisotopic mass)
#        polyethylene glycol (PEG), repeat unit of C2H4O1
            polymer <- "PEG"
            const <- 44/44.0262
            const_mz <- 44.0
#        polyproplyene glycol (PPG), repeat unit of C3H6O1
#            polymer <- "PPG"
#            const <- 58/58.0419
#            const_mz <- 58.0
#        custom
#            polymer <- "custom"
#            const <- integer mass/monoisotopic mass
#            const_mz <- interger mass of unit repeat
```

## KMF Calculations

```r
# PERFORM: run cell
#    convert m/z values from IUPAC scale to kendrick mass scale
#    check the m/z before and after scaling
    kend <- vec[,1]*const
    head(vec[1:5])
    head(kend[1:5])

#    calculate the Kendrick mass defect by subtracting the kendrick exact m/z from the kendrick nominal m/z
#    check mass defect values
    msdefect <- round(kend)-kend
    head(msdefect[1:5])
```

```r
# PERFORM: run cell
#    create new matrix with information included (and rename columns)
#    check first couple of rows in the new matrix
    mat <- cbind(vec, msdefect, kend, round(kend))
    colnames(mat) <- c("m/z","RT","msdefect","kend","nom_kend")
    head(mat)
```

```r
# PERFORM: Set Initial Kendrick Mass Filter Criteria : SET MAJOR PARAMETERS!, run cell

#-----------------------------------------------------------------------    
                        ###IMPORTANT PARAMETER###
#    Change this value iteratively until we understand the optimal value
    mass_defect_parameter <- 0.005
    retention_time <- 1.0
#-----------------------------------------------------------------------

#    calculate Kendrick mass defect
    v1 <- abs(mat[cb[1,],3]-mat[cb[2,],3])<= mass_defect_parameter
#    the right most value corresponds to the nominal m/z difference of the PEG repeat unit
    v2 <- round(abs(mat[cb[1,],4]-mat[cb[2,],4]))==const_mz
#    retention time criterion (mins)
    v4 <- abs(mat[cb[1,],2]-mat[cb[2,],2])<=retention_time

#    determine which values (v3) match the initial criteria
#        v1 = a sufficiently similar mass defeect
#        v2 = the integer m/z difference (set in prior cell, "const_mz")
#        v4 = the RT difference must be less than defined number (min)
    v3 <- v1 & v2 & v4

#    calculate which features in the cb matrix meet filter criteria
    vpos <- unique(as.vector(cb[,v3]))

#    create a matrix with mass/retention time and position for kendrick filtering
    matpos <- cbind(mat[vpos,], vpos)
    head(matpos)
dim(matpos)
```

```r
# PERFORM: run cell
#    "add" back MS1 features based on a minimum number of observations in the data
#        build all relationship of mass differences in filtered MS1 variables
#            note: this step may take some time
    gr <- ftM2graphNEL(as.matrix(t(cb[,v3])), edgemode = "undirected")
    conn <- connComp(gr)
    connLength <- unlist(lapply(conn, length))
#    create a matrix with node degree
    fpos <- cbind(names(degree(gr)), degree(gr))
    sz <- sapply(fpos[,1], function(y) connLength[which(unlist(lapply(conn, function(x) sum(x==y)))==1)])
    fpos <- cbind(fpos, sz)
    #head(fpos)
```

```r
# PERFORM: set mininum number of connections between MS1 features (node degree), run cell 
#    filtering only the nodes with "n" number of connection
#-----------------------------------------------------------------------  
                        ###IMPORTANT PARAMETER###
#    how many connections would you like to set as the minimum number of connections to be considered
        connections_min_number <- 1
#----------------------------------------------------------------------- 
    ffpos <- as.numeric(fpos[as.numeric(fpos[,3])> connections_min_number, 1])
    length(ffpos)
```

```r
# PERFORM: run cell
#    create new matrix with all information relevant to MS features which meet filtering criterion
    mat2 <- cbind(mat[cb[1,v3], 3:4], mat[cb[2,v3], 3:4])
    mat2 <- cbind(mat2, mat2[,1]-mat2[,3], mat2[,2]-mat2[,4])
    colnames(mat2) <- c("msdefect","kend","msdefect","kend","massdefect_difference","mass_difference")
    head(mat2)
```

```r
# PERFORM: run cell
#    create matrix with MS features the meet criteria removed (before reintroduction of min # of connections)
    matdata_Kendrickfiltered <- mat2data[,-c(vpos+1)]
    head(matdata_Kendrickfiltered)
    dim(matdata_Kendrickfiltered)

#    create matrix with MS features the meet criteria removed (Final filtering values)
    matdata_Kendrickfiltered_final <- mat2data[,-c(ffpos+1)]
    head(matdata_Kendrickfiltered_final)
    dim(matdata_Kendrickfiltered_final)
```

```r
# SUMMARY, run cell
    print('Summary', quote = FALSE)
        print(paste('Polymer Filtered - ', polymer), quote = FALSE)
        print(paste('Parameter: mass defect - ', mass_defect_parameter), quote = FALSE)
        print(paste('Parameter: retention time - ', retention_time), quote = FALSE)
        print(paste('Parameter: node degree (minimum # of connection) - ', connections_min_number), quote = FALSE)
        print(paste('number of MS1 features before filtering - ', length(mat2data[1,])), quote = FALSE)
        print(paste('number of MS1 features after filtering (before reincluding based on minimum # of connections) - ', length(mat2data[1,-c(vpos+1)])), quote = FALSE)
        print(paste('number of MS features after filtering (with reincluded based on minimum # of connections) - ', length(mat2data[1,-c(ffpos+1)])), quote = FALSE)
```
